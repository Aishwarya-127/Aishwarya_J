```python
from transformers import pipeline

# Load a translation pipeline (English to French)
translator = pipeline("translation_en_to_fr", model="t5-small")

def translate_text():
    print("Simple Language Translator (English to French)")
    print("Type 'exit' to quit.\n")

    while True:
        text = input("Enter English text: ")
        if text.lower() == "exit":
            print("Goodbye!")
            break

        translation = translator(text, max_length=100)
        translated_text = translation[0]['translation_text']
        print(f"French Translation: {translated_text}\n")

# Run the translator
translate_text()
```
