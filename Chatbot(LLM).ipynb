```python
!pip install transformers

from transformers import pipeline

# Load a very small instruction-following model
chatbot = pipeline("text2text-generation", model="google/flan-t5-small")

def fast_chatbot():
    print("Welcome to Fast Chatbot! Type 'exit' to quit.\n")
    while True:
        user_input = input("You: ")
        if user_input.lower() == "exit":
            print("Goodbye!")
            break

        response = chatbot(user_input, max_length=128)
        reply = response[0]['generated_text']
        print(f"Bot: {reply}\n")

# Run the chatbot
fast_chatbot()
```
